```python
import requests

def baixar_arquivo_json(url, nome_arquivo):
    # Faz o pedido HTTP GET para obter o conteúdo do arquivo JSON
    response = requests.get(url)
    
    # Verifica se a solicitação foi bem-sucedida (status code 200)
    if response.status_code == 200:
        # Abre o arquivo local para escrita binária
        with open(nome_arquivo, 'wb') as arquivo:
            # Escreve o conteúdo do arquivo recebido na resposta HTTP
            arquivo.write(response.content)
        print(f"Arquivo JSON baixado com sucesso: {nome_arquivo}")
    else:
        print(f"Erro ao baixar o arquivo JSON. Status code: {response.status_code}")

# URL do arquivo JSON que você deseja baixar
url_json = "https://jsonplaceholder.typicode.com/posts/1"

# Nome do arquivo local onde você deseja salvar o arquivo JSON
nome_arquivo_local = "post_1.json"

# Chamada da função para baixar o arquivo JSON
baixar_arquivo_json(url_json, nome_arquivo_local)
```

```python
import requests
from bs4 import BeautifulSoup
import json
import nltk
from nltk.corpus import stopwords
from urllib.parse import urljoin
from collections import defaultdict


class Coletor:
    def __init__(self):
        self.urls = []
        self.objects_url = []

    def adicionar_url(self, url):
        self.urls.append(url)

    def extrair_informacoes(self, profundidade=1, largura=True, limite_paginas=10):
        if largura:
            self.busca_largura(profundidade, limite_paginas)
        else:
            self.busca_profundidade(profundidade, limite_paginas)

    def busca_largura(self, profundidade_maxima, limite_paginas):
        visitados = set()
        fila = [(url, 1) for url in self.urls]

        while fila and len(self.objects_url) < limite_paginas:
            url, profundidade = fila.pop(0)
            if url not in visitados and profundidade <= profundidade_maxima:
                visitados.add(url)
                try:
                    resposta = requests.get(url, verify=False)  # Desabilitar verificação SSL
                    print(f"Visitando: {url}")
                    if resposta.status_code == 200:
                        soup = BeautifulSoup(resposta.text, 'html.parser')
                        self.objects_url.append(Url(soup, url))
                        links = [urljoin(url, link['href']) for link in soup.find_all('a', href=True)]
                        for link in links:
                            if link not in visitados:
                                fila.append((link, profundidade + 1))
                    else:
                        print(f"Falha ao carregar a página {url}: {resposta.status_code}")
                except requests.exceptions.RequestException as e:
                    print(f"Erro ao fazer a solicitação para {url}: {e}")

    def busca_profundidade(self, profundidade_maxima, limite_paginas):
        visitados = set()

        def dfs(url, profundidade):
            if len(self.objects_url) >= limite_paginas:
                return
            if profundidade > profundidade_maxima:
                return
            visitados.add(url)
            try:
                resposta = requests.get(url, verify=False)  # Desabilitar verificação SSL
                print(f"Visitando: {url}")
                if resposta.status_code == 200:
                    soup = BeautifulSoup(resposta.text, 'html.parser')
                    self.objects_url.append(Url(soup, url))
                    links = [urljoin(url, link['href']) for link in soup.find_all('a', href=True)]
                    for link in links:
                        if link not in visitados:
                            dfs(link, profundidade + 1)
                else:
                    print(f"Falha ao carregar a página {url}: {resposta.status_code}")
            except requests.exceptions.RequestException as e:
                print(f"Erro ao fazer a solicitação para {url}: {e}")

        for url in self.urls:
            dfs(url, 1)


class Url:
    def __init__(self, soup, url):
        self.url = url
        self.pagina = soup.title.string if soup.title else None
        self.conteudo = [elemento.text for elemento in soup.find_all(['h1', 'h2', 'h3', 'p'])]
        self.links = [urljoin(url, link['href']) for link in soup.find_all('a', href=True)]


class Indexador:
    def __init__(self):
        self.indice_invertido = defaultdict(dict)
        self.stop_words = set(stopwords.words('portuguese'))

    def indexar(self, url_obj):
        if url_obj.pagina:
            tokens = [token.lower() for token in nltk.word_tokenize(url_obj.pagina)
                      if token.isalnum() and token.lower() not in self.stop_words]
            for token in tokens:
                self.indice_invertido[token][url_obj.url] = self.indice_invertido[token].get(url_obj.url, 0) + 1

    def salvar_indice(self, nome_arquivo):
        with open(nome_arquivo, 'w') as arquivo:
            json.dump(self.indice_invertido, arquivo, indent=4)


if __name__ == "__main__":
    urls = [
        "https://www.gov.br/inep/pt-br/areas-de-atuacao/avaliacao-e-exames-educacionais/enem",
        "https://ouropreto.ifmg.edu.br/ouropreto",
        "https://brasilescola.uol.com.br"
    ]

    coletor = Coletor()
    indexador = Indexador()

    for url in urls:
        coletor.adicionar_url(url)

    coletor.extrair_informacoes(profundidade=2, largura=True, limite_paginas=10)

    for url_obj in coletor.objects_url:
        indexador.indexar(url_obj)

    indexador.salvar_indice("indice_invertido.json")
```

```python
import requests
from bs4 import BeautifulSoup
import json
from urllib.parse import urljoin
from collections import defaultdict
from collections import Counter
import chardet
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

class Coletor:
    def __init__(self):
        self.urls = []
        self.objects_url = []

    def adicionar_url(self, url):
        self.urls.append(url)

    def extrair_informacoes(self, profundidade=1, largura=True, limite_paginas=10):
        if largura:
            self.busca_largura(profundidade, limite_paginas)
        else:
            self.busca_profundidade(profundidade, limite_paginas)

    def busca_largura(self, profundidade_maxima, limite_paginas):
        visitados = set()
        fila = [(url, 1) for url in self.urls]

        while fila and len(self.objects_url) < limite_paginas:
            url, profundidade = fila.pop(0)
            if url not in visitados and profundidade <= profundidade_maxima:
                visitados.add(url)
                try:
                    resposta = requests.get(url, verify=False)  # Desabilitar verificação SSL
                    print(f"Visitando: {url}")
                    if resposta.status_code == 200:
                        soup = BeautifulSoup(resposta.content, 'html.parser')
                        self.objects_url.append(Url(soup, url))
                        links = [urljoin(url, link['href']) for link in soup.find_all('a', href=True)]
                        for link in links:
                            if link not in visitados:
                                fila.append((link, profundidade + 1))
                    else:
                        print(f"Falha ao carregar a página {url}: {resposta.status_code}")
                except requests.exceptions.RequestException as e:
                    print(f"Erro ao fazer a solicitação para {url}: {e}")

    def busca_profundidade(self, profundidade_maxima, limite_paginas):
        visitados = set()

        def dfs(url, profundidade):
            if len(self.objects_url) >= limite_paginas:
                return
            if profundidade > profundidade_maxima:
                return
            visitados.add(url)
            try:
                resposta = requests.get(url, verify=False)  # Desabilitar verificação SSL
                print(f"Visitando: {url}")
                if resposta.status_code == 200:
                    soup = BeautifulSoup(resposta.content, 'html.parser')
                    self.objects_url.append(Url(soup, url))
                    links = [urljoin(url, link['href']) for link in soup.find_all('a', href=True)]
                    for link in links:
                        if link not in visitados:
                            dfs(link, profundidade + 1)
                else:
                    print(f"Falha ao carregar a página {url}: {resposta.status_code}")
            except requests.exceptions.RequestException as e:
                print(f"Erro ao fazer a solicitação para {url}: {e}")

        for url in self.urls:
            dfs(url, 1)


class Url:
    def __init__(self, soup, url):
        self.url = url
        self.pagina = soup.title.string if soup.title else ''
        self.conteudo = [elemento.text for elemento in soup.find_all(['h1', 'h2', 'h3', 'p'])]
        self.links = [urljoin(url, link['href']) for link in soup.find_all('a', href=True)]


class Indexador:
    def __init__(self):
        self.indice_invertido = defaultdict(dict)
        self.stop_words = set(stopwords.words('portuguese'))

    def detectar_codificacao(self, texto):
        deteccao = chardet.detect(texto)
        return deteccao['encoding']

    def indexar(self, url_obj):
        if url_obj.pagina:
            codificacao = self.detectar_codificacao(url_obj.pagina.encode())
            pagina_decodificada = url_obj.pagina.encode().decode(codificacao, 'ignore')
            tokens = word_tokenize(pagina_decodificada.lower())
            tokens = [token for token in tokens if token.isalnum() and token not in self.stop_words]
            freq_tokens = Counter(tokens)
            for token, freq in freq_tokens.items():
                self.indice_invertido[token][url_obj.url] = freq

            conteudo_texto = ' '.join(url_obj.conteudo)
            codificacao_conteudo = self.detectar_codificacao(conteudo_texto.encode())
            conteudo_decodificado = conteudo_texto.encode().decode(codificacao_conteudo, 'ignore')
            tokens_conteudo = word_tokenize(conteudo_decodificado.lower())
            tokens_conteudo = [token for token in tokens_conteudo if token.isalnum() and token not in self.stop_words]
            freq_tokens_conteudo = Counter(tokens_conteudo)
            for token, freq in freq_tokens_conteudo.items():
                self.indice_invertido[token][url_obj.url] = freq

    def salvar_indice(self, nome_arquivo):
        with open(nome_arquivo, 'w') as arquivo:
            json.dump(self.indice_invertido, arquivo, indent=4)

if __name__ == "__main__":
    urls = [
        "https://www.gov.br/inep/pt-br/areas-de-atuacao/avaliacao-e-exames-educacionais/enem",
        "https://ouropreto.ifmg.edu.br/ouropreto",
        "https://brasilescola.uol.com.br"
    ]

    coletor = Coletor()
    indexador = Indexador()

    for url in urls:
        coletor.adicionar_url(url)

    coletor.extrair_informacoes(profundidade=2, largura=True, limite_paginas=10)

    for url_obj in coletor.objects_url:
        indexador.indexar(url_obj)

    indexador.salvar_indice("indice_invertido.json")
```

```python
# Definição fictícia do índice invertido
indice_invertido = {
    "educação": {"documento1": 2, "documento2": 1},
    "enem": {"documento1": 3, "documento3": 2},
    "vestibular": {"documento2": 1, "documento3": 1},
    "alunos": {"documento3": 2}
}

def pesquisa_booleana(consulta, indice_invertido):
    termos = consulta.split()
    operadores = set(["AND", "OR", "NOT"])
    resultados = set()
    operador = "AND"
    stack = []
    
    for termo in termos:
        if termo == "(":
            stack.append(resultados)
            stack.append(operador)
            resultados = set()
            operador = "AND"
        elif termo == ")":
            resultados_antigos = stack.pop()
            operador = stack.pop()
            if operador == "AND":
                resultados &= resultados_antigos
            elif operador == "OR":
                resultados |= resultados_antigos
            elif operador == "NOT":
                resultados -= resultados_antigos
        elif termo in operadores:
            operador = termo
        else:
            documentos = set(indice_invertido.get(termo, {}).keys())
            if operador == "AND":
                resultados &= documentos
            elif operador == "OR":
                resultados |= documentos
            elif operador == "NOT":
                resultados -= documentos
    
    return resultados

# Uso
consulta = "educação AND (enem OR vestibular) NOT alunos"
resultados = pesquisa_booleana(consulta, indice_invertido)
print("Documentos relevantes:", resultados)
```

3 - Ranking de Resultados

```python
def ranking_resultados(termos, indice_invertido):
    resultados = defaultdict(float)
    for termo in termos:
        if termo in indice_invertido:
            for documento, freq in indice_invertido[termo].items():
                resultados[documento] += freq
    return dict(sorted(resultados.items(), key=lambda x: x[1], reverse=True))

# Uso
termos = ["educação", "enem"]
resultados_ranking = ranking_resultados(termos, indice_invertido)
print("Documentos classificados por relevância:", resultados_ranking)
```

4 - Análise de sentimento 

```python
def analise_sentimento(texto, dicionario_sentimento):
    # Divide o texto em palavras
    palavras = texto.lower().split()
    
    # Inicializa a pontuação do sentimento
    pontuacao_sentimento = 0
    
    # Calcula a pontuação do sentimento para cada palavra no texto
    for palavra in palavras:
        if palavra in dicionario_sentimento:
            pontuacao_sentimento += dicionario_sentimento[palavra]
    
    # Retorna o resultado da análise de sentimento
    if pontuacao_sentimento > 0:
        return "Positivo"
    elif pontuacao_sentimento < 0:
        return "Negativo"
    else:
        return "Neutro"

# Dicionário de palavras associadas ao sentimento
dicionario_sentimento = {
    "bom": 1,
    "ruim": -1,
    "excelente": 2,
    "terrível": -2,
    "normal": 0,
    "feliz": 1,
    "triste": -1
}

# Texto de exemplo
texto_exemplo = "Eu me sinto muito feliz hoje, o tempo está bom e o dia está excelente."

# Realiza a análise de sentimento
sentimento = analise_sentimento(texto_exemplo, dicionario_sentimento)
print("Sentimento:", sentimento)
```

A classe Indexador possui métodos para indexar o conteúdo das páginas e os links coletados.
O método indexar_pagina divide o texto da página em palavras e adiciona cada palavra ao índice invertido junto com a URL da página onde a palavra ocorre.
O método indexar_links adiciona os links coletados ao índice invertido, associando cada link com a URL da página onde foi encontrado.
O método mostrar_indice exibe o índice invertido na saída padrão.

```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import urljoin

class Indexador:
    def __init__(self):
        self.indice_invertido = {}

    def indexar_pagina(self, url, conteudo):
        soup = BeautifulSoup(conteudo, 'html.parser')
        texto_pagina = soup.get_text()
        palavras = texto_pagina.split()
        for palavra in palavras:
            if palavra not in self.indice_invertido:
                self.indice_invertido[palavra] = [url]
            else:
                if url not in self.indice_invertido[palavra]:
                    self.indice_invertido[palavra].append(url)

    def indexar_links(self, url, links):
        for link in links:
            if link not in self.indice_invertido:
                self.indice_invertido[link] = [url]
            else:
                if url not in self.indice_invertido[link]:
                    self.indice_invertido[link].append(url)

    def mostrar_indice(self):
        for termo, ocorrencias in self.indice_invertido.items():
            print(f"{termo}: {ocorrencias}")

# Exemplo de uso
if __name__ == "__main__":
    urls = [
        "https://www.gov.br/inep/pt-br/areas-de-atuacao/avaliacao-e-exames-educacionais/enem",
        "https://ouropreto.ifmg.edu.br/ouropreto",
        "https://brasilescola.uol.com.br"
    ]

    indexador = Indexador()

    for url in urls:
        resposta = requests.get(url)
        if resposta.status_code == 200:
            conteudo_pagina = resposta.text
            indexador.indexar_pagina(url, conteudo_pagina)
            soup = BeautifulSoup(conteudo_pagina, 'html.parser')
            links = [urljoin(url, link.get('href')) for link in soup.find_all('a', href=True)]
            indexador.indexar_links(url, links)

    indexador.mostrar_indice()
```

```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import urljoin
from collections import Counter
import math

class IndexadorTFIDF:
    def __init__(self):
        self.indice_invertido = {}
        self.num_documentos = 0

    def indexar_pagina(self, url, conteudo):
        self.num_documentos += 1
        soup = BeautifulSoup(conteudo, 'html.parser')
        texto_pagina = soup.get_text()
        palavras = [palavra.lower() for palavra in texto_pagina.split()]
        tf = Counter(palavras)
        for termo, frequencia in tf.items():
            if termo not in self.indice_invertido:
                self.indice_invertido[termo] = {url: frequencia}
            else:
                self.indice_invertido[termo][url] = frequencia

    def calcular_tfidf(self):
        for termo, ocorrencias in self.indice_invertido.items():
            idf = math.log(self.num_documentos / len(ocorrencias))
            for url, tf in ocorrencias.items():
                tfidf = tf * idf
                self.indice_invertido[termo][url] = tfidf

    def mostrar_indice(self):
        for termo, ocorrencias in self.indice_invertido.items():
            print(f"{termo}: {ocorrencias}")

# Exemplo de uso
if __name__ == "__main__":
    urls = [
        "https://www.gov.br/inep/pt-br/areas-de-atuacao/avaliacao-e-exames-educacionais/enem",
        "https://ouropreto.ifmg.edu.br/ouropreto",
        "https://brasilescola.uol.com.br"
    ]

    indexador = IndexadorTFIDF()

    for url in urls:
        resposta = requests.get(url)
        if resposta.status_code == 200:
            conteudo_pagina = resposta.text
            indexador.indexar_pagina(url, conteudo_pagina)

    indexador.calcular_tfidf()
    indexador.mostrar_indice()
```
