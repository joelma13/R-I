```python
import requests

def baixar_arquivo_json(url, nome_arquivo):
    # Faz o pedido HTTP GET para obter o conteúdo do arquivo JSON
    response = requests.get(url)
    
    # Verifica se a solicitação foi bem-sucedida (status code 200)
    if response.status_code == 200:
        # Abre o arquivo local para escrita binária
        with open(nome_arquivo, 'wb') as arquivo:
            # Escreve o conteúdo do arquivo recebido na resposta HTTP
            arquivo.write(response.content)
        print(f"Arquivo JSON baixado com sucesso: {nome_arquivo}")
    else:
        print(f"Erro ao baixar o arquivo JSON. Status code: {response.status_code}")

# URL do arquivo JSON que você deseja baixar
url_json = "https://jsonplaceholder.typicode.com/posts/1"

# Nome do arquivo local onde você deseja salvar o arquivo JSON
nome_arquivo_local = "post_1.json"

# Chamada da função para baixar o arquivo JSON
baixar_arquivo_json(url_json, nome_arquivo_local)
```

```python
import requests

class Coletor:
    def __init__(self):
        self.urls = []

    def addUrl(self, url):
        self.urls.append(url)

    def test_connection(self):
        for url in self.urls:
            if self.is_accessible(url):
                print(f"Conexão bem-sucedida com: {url}")
            else:
                print(f"Falha ao conectar-se a: {url}")

    def is_accessible(self, url):
        try:
            response = requests.get(url)
            return response.status_code == 200
        except Exception as e:
            print(f"Erro ao acessar {url}: {e}")
            return False

# Lista de URLs para testar a conexão
urls = [
    "https://www.gov.br/inep/pt-br/areas-de-atuacao/avaliacao-e-exames-educacionais/enem",
    "https://ouropreto.ifmg.edu.br/ouropreto",
    "https://brasilescola.uol.com.br"
]

# Criando um coletor
coletor = Coletor()

# Adicionando URLs ao coletor
for url in urls:
    coletor.addUrl(url)

# Testando a conexão com as URLs
coletor.test_connection()
```

```python
import requests
from bs4 import BeautifulSoup

class Coletor:
    def __init__(self):
        self.urls = []

    def addUrl(self, url):
        self.urls.append(url)

    def extract_information(self):
        for url in self.urls:
            response = requests.get(url)
            if response.status_code == 200:
                soup = BeautifulSoup(response.text, 'html.parser')
                
                # Extraindo e imprimindo o título da página
                title = soup.title.text.strip()
                print(f"Título da página ({url}): {title}")
                
                # Extraindo e imprimindo tabelas da página
                print("Tabelas da página:")
                tables = soup.find_all('table')
                for table in tables:
                    for row in table.find_all('tr'):
                        cells = row.find_all('td')
                        if cells:
                            row_data = [cell.text.strip() for cell in cells]
                            print(row_data)
                
                print("---------------")
            else:
                print(f"Falha ao carregar a página: {url}")

# Lista de URLs para testar a conexão
urls = [
    "https://www.gov.br/inep/pt-br/areas-de-atuacao/avaliacao-e-exames-educacionais/enem",
    "https://ouropreto.ifmg.edu.br/ouropreto",
    "https://brasilescola.uol.com.br"
]

# Criando um coletor
coletor = Coletor()

# Adicionando URLs ao coletor
for url in urls:
    coletor.addUrl(url)

# Extraindo informações das páginas
coletor.extract_information()
```
